```python
#Welcome Function which will run everytime when you buy a product
def welcome():
    print("--------------------------------------------------------------")  
    print("\nCategory::\n\nA]Food Products\nB]Electronics\nC]Clothing\nD]Daily Products\n")
    
    x=input("Choose the Category (Eg: A,B,C,D) ::")  # Filter by Categories for easy shopping
    if x.upper()=="A":
        ccategory="Food Products"
    elif x.upper()=="B":
        ccategory="Electronics"
    elif x.upper()=="C":
        ccategory="Clothing"
    elif x.upper()=="D":
        ccategory="Daily Products"
     
    # Shows the filtered Data 
    print("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++") 
    print("BarCode  ||Name Of Product||Quantity||Price(RS)||Discount(%)")
    print("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++") 
    
    for x in record[ccategory].keys():
        print(x," ",record[ccategory][x]['name'].center(17,' '),"    ",record[ccategory][x]['quantity'],"     ",record[ccategory][x]['price'],"     ",record[ccategory][x]['discount'])
    print("--------------------------------------------------------------")     
    return ccategory
    
```

```python
#Showing all the order in bill format with name of customers and time 
def show_bill(bill,cname):
    
    t=0
    print(" Bill ".center(75,'#'))
    print("\n Mr./Mrs.",cname,"Time: ".rjust(40," "),time.ctime(),"\n")
    
    billt[cname]= bill
    
    print("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++") 
    print(" BarCode  ||Purchased Product||Quantity||Price(RS)||Discount(%)||Total")
    print("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
    
    for x in bill.keys():
        print(" ",x," ",bill[x]['name'].center(20,' '),"  ",bill[x]['quantity'],"      ",bill[x]['price'],"       ",bill[x]['discount'],"    ",bill[x]['total'])
        print("------------------------------------------------------------------------")
        t+= bill[x]['total']
        
    print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
    print(" Total Amount ::".rjust(62,' '),t)
    print("\n#####################################################################")
    

    
```

```python
# Update both the Files storage and sales
def update(billt):
    sales.update(billt)
    
    file=open("storage.json","w")
    txt=json.dumps(record)
    file.write(txt)
    file.close()
    
    file=open("sales.json","w")
    sls=json.dumps(sales)
    file.write(sls)
    file.close()
    
```

```python
print(" Welcome  ".center(70,'#'))
cname=input("\nEnter your Name ::")

total=0
b=True
while b==True:
    
    ccategory=welcome()
    
    cbar_code=input("Enter the product bar code ::")
    cquantity=int(input("Enter the quantity of product ::"))
    
    if (cquantity > record[ccategory][cbar_code]['quantity'] or cquantity<= 0):
        print("Product requested quantity is out of limit.\nPlease Try Again!!!")
        continue
        
    else:
        record[ccategory][cbar_code]['quantity']-= cquantity    #Removes the customers quanity from  Storage quantity
        total= cquantity * record[ccategory][cbar_code]['price'] * (100 - record[ccategory][cbar_code]['discount']) / 100
        
        # Show the current order
        print("---------------------------------------------------------------------")
        print("\nName of Product : ",record[ccategory][cbar_code]['name'],"\nQuantity : ",cquantity,"\nPrice : ","Rs.",record[ccategory][cbar_code]['price'],"\nDiscount : ",record[ccategory][cbar_code]['discount'],"%","\nTotal amount : ","Rs.",total)
        print("---------------------------------------------------------------------")   
        bill[cbar_code]= { 'name':record[ccategory][cbar_code]['name'], 'quantity': cquantity, 'price':record[ccategory][cbar_code]['price'],'discount':record[ccategory][cbar_code]['discount'],'total': total}
        print ("**Order Added to Cart**")
        
        print("---------------------------------------------------------------------")
        k= input("Do you want to buy another Product (Y/N) ::")
        print("---------------------------------------------------------------------")
        
        if k.lower() == 'n':
            print("**Generating Bill**\n")
            show_bill(bill,cname)
            print("**Procceding to Pay**\n")
            print("---------------------------------------------------------------------")
            
            print("A]Pay and Exit\nB]Rebuy everything\nC]Exit\n")
            j= input("Choose (Eg: A,B,C):: ")
            
            if j.lower()=="a":
                print("\n**Payement Paid** ")
                update(billt)
                print("**Exiting from IMS**")
                b= False
            elif j.lower() == "b":
                print ("**All the product orders in cart will be removed**")
                continue
            elif j.lower() == "c":
                print("**Exiting from IMS**")
                b= False
                
                       
            
```
